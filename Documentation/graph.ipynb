```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import FancyBboxPatch
import numpy as np

fig, ax = plt.subplots(1, 1, figsize=(10, 12))
ax.set_xlim(0, 10)
ax.set_ylim(0, 14)
ax.axis('off')

# Define colors
primary_color = '#7CB342'    # Green
secondary_color = '#2C2C2C'  # Black
accent_color = '#4CAF50'     # Accent green for arrows
light_color = '#66BB6A'      # Light green

# Pipeline stages with descriptions
stages = [
    ("Raw Data Collection", "Visitor logs, Weather APIs, Holiday calendars, Crew schedules", 12),
    ("Data Preprocessing", "Cleaning, validation, missing value handling", 10.5),
    ("Feature Engineering", "Temporal encoding, lagged features, rolling windows", 9),
    ("Model Training", "XGBoost for 7 visitor segments + crew size classification", 7.5),
    ("Model Validation", "Cross-validation, performance metrics", 6),
    ("Prediction Generation", "7/14/21-day visitor forecasts + crew optimization", 4.5),
    ("Dashboard Integration", "Real-time predictions with user-friendly interface", 3)
]

# Draw pipeline boxes and arrows
for i, (title, description, y_pos) in enumerate(stages):
    # Main box
    box = FancyBboxPatch((1, y_pos-0.4), 8, 0.8, 
                        boxstyle="round,pad=0.1", 
                        facecolor=primary_color if i % 2 == 0 else secondary_color,
                        edgecolor='white', linewidth=2)
    ax.add_patch(box)
    
    # Title text
    ax.text(5, y_pos+0.1, title, fontsize=12, fontweight='bold', 
           ha='center', va='center', color='white')
    
    # Description text
    ax.text(5, y_pos-0.2, description, fontsize=9, 
           ha='center', va='center', color='white', style='italic')
    
    # Arrow to next stage
    if i < len(stages) - 1:
        ax.arrow(5, y_pos-0.5, 0, -0.6, head_width=0.2, head_length=0.1, 
                fc=accent_color, ec=accent_color, linewidth=2)

# Add title
ax.text(5, 13.5, "AA_NEMO ML Pipeline", fontsize=16, fontweight='bold', 
       ha='center', va='center')

plt.tight_layout()
plt.show()
```
